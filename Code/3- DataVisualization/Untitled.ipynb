```python
import matplotlib.pyplot as plt
import numpy as np

# Precision values for Support Vector Machine (SVM) and other models
# SVM metrics from the provided data
high_precision_svm = 0.99
low_precision_svm = 0.00
moderate_precision_svm = 0.59

# Average precision for SVM
average_precision_svm = (high_precision_svm + low_precision_svm + moderate_precision_svm) / 3

# Precision values for Gradient Boosting, Logistic Regression, Naive Bayes, Random Forest, and SVM
high_precision = [0.98, 0.92, 1.00, 0.99, high_precision_svm]
low_precision = [0.94, 0.00, 0.43, 0.94, low_precision_svm]
moderate_precision = [0.86, 0.47, 0.37, 0.86, moderate_precision_svm]

# Calculate average precision for each model
average_precision = [(h + l + m) / 3 for h, l, m in zip(high_precision, low_precision, moderate_precision)]

# Models and other performance metrics
models = ['Gradient Boosting', 'Logistic Regression', 'Naive Bayes', 'Random Forest', 'Support Vector Machine']
accuracy = [0.9675, 0.8648, 0.8495, 0.9713, 0.9168]  # Including SVM
precision = [np.mean([0.98, 0.94, 0.86]), average_precision[0], average_precision[2], average_precision[3], average_precision_svm]  # Average precision for SVM
recall = [0.98, 0.97, 0.90, 0.98, 0.97]    # Recall for SVM is 0.97 (High)
f1_score = [0.98, 0.94, 0.95, 0.99, 0.98]  # F1-Score for SVM is 0.98 (High)

# Creating the plot
x = np.arange(len(models))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 7))

# Creating bars for each metric
rects1 = ax.bar(x - width*1.5, accuracy, width, label='Accuracy', color='tab:orange')
rects2 = ax.bar(x - width*0.5, precision, width, label='Precision', color='tab:red')
rects3 = ax.bar(x + width*0.5, recall, width, label='Recall', color='tab:blue')
rects4 = ax.bar(x + width*1.5, f1_score, width, label='F1 Score', color='tab:purple')

# Adding text labels, title, and axes labels
ax.set_xlabel('Models', fontsize=14)
ax.set_ylabel('Scores', fontsize=14)
ax.set_title('Model Performance Comparison', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=12)
ax.legend(fontsize=12)

# Annotate with the actual values on top of the bars
def add_values(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=12)

add_values(rects1)
add_values(rects2)
add_values(rects3)
add_values(rects4)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the models and their evaluation metrics
models = ['Gradient Boosting', 'Logistic Regression', 'Naive Bayes', 'Random Forest', 'Support Vector Machine']
accuracy = [0.9675, 0.8648, 0.8495, 0.9713, 0.9168]
precision = [0.98, 0.92, 1.00, 0.99, 0.99]  # Example values for High Precision
recall = [0.98, 0.97, 0.90, 0.98, 0.97]    # Recall for High
f1_score = [0.98, 0.94, 0.95, 0.99, 0.98]  # F1-Score for High

# Create a matrix of metrics for visualization
metrics = np.array([accuracy, precision, recall, f1_score])

# Create a DataFrame for better labeling
import pandas as pd
metrics_df = pd.DataFrame(metrics, index=['Accuracy', 'Precision', 'Recall', 'F1-Score'], columns=models)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(metrics_df, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Model Evaluation Metrics")
plt.ylabel('Metrics')
plt.xlabel('Models')
plt.show()
```
